```python
# ğŸ“Œ 1ï¸âƒ£ Import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Regression models
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor

# ğŸ“Œ 2ï¸âƒ£ Load dataset
df = pd.read_csv("dirty_house_prices.csv")   # â† use your dataset name here
print("Initial shape:", df.shape)

# ğŸ“Œ 3ï¸âƒ£ Define Features (X) and Target (y)
X = df.drop('price', axis=1)
y = df['price']

# ğŸ“Œ 4ï¸âƒ£ Identify Numerical and Categorical Columns
num_features = X.select_dtypes(exclude='object').columns
cat_features = X.select_dtypes(include='object').columns

# ğŸ“Œ 5ï¸âƒ£ Build Preprocessing Pipelines
num_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean'))
])

cat_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_transformer, num_features),
        ('cat', cat_transformer, cat_features)
    ]
)

# ğŸ“Œ 6ï¸âƒ£ Preprocess Features
X_processed = preprocessor.fit_transform(X)

# ğŸ“Œ 7ï¸âƒ£ Train-Test Split
x_train, x_test, y_train, y_test = train_test_split(
    X_processed, y, test_size=0.2, random_state=42
)

# ğŸ“Œ 8ï¸âƒ£ Evaluation Function
def evaluatefunction(true, pred):
    mse = mean_squared_error(true, pred)
    mae = mean_absolute_error(true, pred)
    r2 = r2_score(true, pred)
    return mse, mae, r2

# ğŸ“Œ 9ï¸âƒ£ Define Models Dictionary
models = {
    "Linear Regression": LinearRegression(),
    "Ridge Regression": Ridge(),
    "Lasso Regression": Lasso(),
    "ElasticNet": ElasticNet(),
    "Decision Tree": DecisionTreeRegressor(),
    "Random Forest": RandomForestRegressor(),
    "Gradient Boosting": GradientBoostingRegressor(),
    "AdaBoost": AdaBoostRegressor(),
    "Support Vector Regressor": SVR(),
    "K-Nearest Neighbors": KNeighborsRegressor()
}

# ğŸ“Œ ğŸ”Ÿ Train & Evaluate Each Model
models_list = []
r2_score_list = []

print("\nğŸ“Š Model Performance Summary:\n")
for name, model in models.items():
    model.fit(x_train, y_train)
    
    # Predictions
    y_train_pred = model.predict(x_train)
    y_test_pred = model.predict(x_test)
    
    # Train metrics
    mse_train, mae_train, r2_train = evaluatefunction(y_train, y_train_pred)
    # Test metrics
    mse_test, mae_test, r2_test = evaluatefunction(y_test, y_test_pred)
    
    models_list.append(name)
    r2_score_list.append(r2_test)
    
    print(f"ğŸ”¸ {name}")
    print(f"   Train â†’ MAE: {mae_train:.4f}, MSE: {mse_train:.4f}, RÂ²: {r2_train:.4f}")
    print(f"   Test  â†’ MAE: {mae_test:.4f}, MSE: {mse_test:.4f}, RÂ²: {r2_test:.4f}\n")

# ğŸ“Œ ğŸ† Best Model
best_idx = r2_score_list.index(max(r2_score_list))
print(f"âœ… Best Model: {models_list[best_idx]} with RÂ² = {max(r2_score_list):.4f}")
```
